```python
A = HDFArchive("dmft_bethe.output.h5",'r') # Open file
for it in range(21): # Use parameter in the analysis
 if it%2: oplot(A['G%i'%it], '-o', mode="I", name='G%i'%it) # Plot every second result
plt.xlim(0,2)
plt.ylim(-1.5,0)
```
